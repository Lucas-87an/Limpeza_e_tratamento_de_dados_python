```python
import pandas as pd
import numpy as np

pd.set_option('display.float_format', '{:.2f}'.format)
```

* DADOS RELACIONADOS A FINANÇAS FUNDOS, LIQUIDEZ E RENTABILIDADE E ETC..

* Carregando e importando dados

```python
# Leitura dos dados
df = pd.read_csv('/content/drive/MyDrive/AD/output/df_fiis.csv', index_col='TICKER')

# Remover colunas desnecessárias
["COTACAO_FECHAMENTO", "VALOR", "TX_GESTAO", "SOMA_YIELD_ANO_CORRENTE", "YELD"]
df = df.drop(
    columns=
     ["COTACAO_FECHAMENTO",
      "VALOR",
      "TX_GESTAO",
      "SOMA_YIELD_ANO_CORRENTE",
      "YELD"
      ]
    )

# Informações gerais dos dados
df.info()
# Visualizar dados
df
```

* Limpeza de dados (Removendo dados inválidos)

```python
# Ajustar valores numéricos
df_filtrado = df.replace("---", np.nan)
df_filtrado[["MEDIA_YIELD_12M", "RENTABILIDADE"]] = df_filtrado[
    ["MEDIA_YIELD_12M", "RENTABILIDADE"]
].astype(float)

# Verificar se os dados estão corretos após correção
df_filtrado.info()
```

```python
# Remover valores NaN
df_filtrado = df_filtrado.dropna()
df_filtrado.isna().sum()

# Visualizar estatisticas
df_filtrado.describe()
```

* Removendo dados nulos

```python
# Remover dados nulos
# "PATRIMONIO_LIQUIDO" e "NUMERO_COTISTA" Não podem ser zero (!!)
filtro_dados_nulos = df_filtrado["PATRIMONIO_LIQUIDO"] > 0

# Filtrando dados
df_filtrado = df_filtrado[filtro_dados_nulos]

# Visualizar estatísticas após correção
df_filtrado.describe()
```

```python
# Percentual de dados removidos
((len(df_filtrado) / len(df))-1) * 100
```

* Removendo dados inconsistentes (outliers)

```python
# Remover outliers (PVP + YIELD + LIQUIDEZ + N COTISTAS)
filtro_outliers = (
    (df_filtrado["PVP"] < 2)
    & (df_filtrado["MEDIA_YIELD_12M"] < 2)
    & (df_filtrado["LIQUIDEZMEDIADIARIA"] > df_filtrado['LIQUIDEZMEDIADIARIA'].median())
    & (df_filtrado["NUMERO_COTISTA"] > df_filtrado['NUMERO_COTISTA'].median())
)
df_filtrado = df_filtrado[filtro_outliers]

# Visualizar estatísticas
df_filtrado.describe()
```
