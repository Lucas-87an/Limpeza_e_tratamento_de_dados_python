DADOS METEOROLÓGICOS NA REGIÃO DO CEARÁ

```python
path_dados = '/content/drive/MyDrive/AD/output/weather_data.csv'

df = pd.read_csv(path_dados, index_col="TS")
display(df)
```

"TS" -> Data, "RH" -> Umidade relativa, "TAMB" -> Temperatura ambiente da região, "GHI" -> radiação da localização

```python
# Visualiza principais estatisticas dos dados
df.describe()
```



*   Filtragem de dados de Umidade relativa



```python
# Definindo limites inferior e superior para detecção de outliers
limite_inferior_rh = 0
limite_superior_rh = 100

# Visualizando limites
print(f"Limite Inferior: {limite_inferior_rh}")
print(f"Limite Superior: {limite_superior_rh}")

# Criando máscara para detecção de outliers
mascara_rh = (df['RH'] < limite_inferior_rh) | (df['RH'] > limite_superior_rh)
```



*   Filtragem de dados de temperatura



```python
# Definindo limites inferior e superior para detecção de outliers
limite_inferior_tamb = df['TAMB'].quantile(0.05) * 0.85
limite_superior_tamb = df['TAMB'].quantile(0.95) * 1.15

# Visualizando limites
print(f"Limite Inferior: {limite_inferior_tamb}")
print(f"Limite Superior: {limite_superior_tamb}")

# Criando máscara para detecção de outliers
mascara_tamb = (df['TAMB'] < limite_inferior_tamb) | (df['TAMB'] > limite_superior_tamb)
```

```python
# Localizando todos os outliers
outliers_total = (mascara_rh | mascara_tamb)
# Filtra os dados removendo os outiliers detectados
df_filtrado = df[~outliers_total]

# Visualizando estatisticas dos dados após filtragem
df_filtrado.describe()
```
